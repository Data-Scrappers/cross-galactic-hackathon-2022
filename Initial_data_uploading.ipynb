```python
#Converting timestamp to date_time format
normal_v0[' Timestamp'] = pd.to_datetime(normal_v0[' Timestamp'])
#Removing gaps
normal_v0.columns = normal_v0.columns.str.replace(' ', '')
```

```python
attacks = pd.read_excel('input_data/SWaT_Dataset_Attack_v0.xlsx', header = 1)
```

```python
attacks[' Timestamp'] = pd.to_datetime(attacks[' Timestamp'])

attacks.columns = attacks.columns.str.replace(' ', '')
```

```python
attacks.head()
```

Attacks file begins when normal_v0 file ends, so we can concat them

```python
dataset = pd.concat([normal_v0,attacks], ignore_index = True)
```

Checking gaps and overall info

```python
dataset.isna().sum().sum()
```

```python
dataset.info()
```

```python
#Replacing dirty values
dataset.loc[dataset['Normal/Attack'] == 'A ttack', 'Normal/Attack'] = 'Attack'
dataset.loc[dataset['Normal/Attack'] == 'Attack', 'Normal/Attack'] = 1
dataset.loc[dataset['Normal/Attack'] == 'Normal', 'Normal/Attack'] = 0
dataset['Normal/Attack']  = dataset['Normal/Attack'].astype('int8')
```

```python
dataset.to_csv('input_data/data.csv')
```

```python
plt.figure(figsize = (20,20))
sns.heatmap(dataset.corr(), vmin = 0);
```

Removing features, which aren't changing during time

```python
drop_list = []
for column in dataset:
    if dataset[column].dtype != 'object' and dataset[column].std() == 0:
        drop_list.append(column)
dataset.drop(drop_list, axis = 1, inplace = True)
```

```python
plt.figure(figsize = (30,30))
sns.heatmap(dataset.corr(), vmin = 0, annot=True, fmt = '.2f');
```
